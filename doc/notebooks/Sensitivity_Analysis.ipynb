```python
import pandas as pd
import seaborn as sns
from ggplot import *
from pandas.tools.plotting import parallel_coordinates
import matplotlib.pyplot as plt
pd.options.display.mpl_style = 'default'

%matplotlib inline
```

```python
data = pd.read_csv('ga_simpledoe_sensitivity.csv')
data = data.dropna(how='all')
data = data.dropna(axis=1, how='all')
#data = data.drop('max_retries', axis=1)
#data = data.drop('retries', axis=1)
data = data.drop('timestamp', axis=1)
data.head()
```

```python

dataroi2 = data.copy()
dataroi3 = data.copy()
dataroi4 = data.copy()
dataroi5 = data.copy()

dataroi2['ga.roi'] = dataroi2['ga.year5Roi']
dataroi2['ga.roiyear'] = 5
dataroi3['ga.roi'] = dataroi3['ga.year10Roi']
dataroi3['ga.roiyear'] = 10
dataroi4['ga.roi'] = dataroi4['ga.year20Roi']
dataroi4['ga.roiyear'] = 20
dataroi5['ga.roi'] = dataroi5['ga.year30Roi']
dataroi5['ga.roiyear'] = 30

data = pd.concat([data, dataroi2, dataroi3, dataroi4, dataroi5])

data.head()
min(data['ga.roiyear'].values)
```

```python
f, ax = sns.plt.subplots(figsize=(10,10))
cmap = sns.blend_palette(["#00008B", "#6A5ACD", "#F0F8FF",
                          "#FFE6F8", "#C71585", "#8B0000"], as_cmap=True)
sns.corrplot(data, annot=False, diag_names=False, cmap=cmap)
ax.grid(False)
```

```python
highRoi = data[data['ga.year5Roi']>0]
sns.lmplot("ga.runSpeed", "ga.year5Roi", highRoi)
```

```python
superhighRoi = data[data['ga.year5Roi']>1000000]

f, ax = sns.plt.subplots(figsize=(12,12))
cmap = sns.blend_palette(["#00008B", "#6A5ACD", "#F0F8FF",
                          "#FFE6F8", "#C71585", "#8B0000"], as_cmap=True)
sns.corrplot(superhighRoi, annot=False, diag_names=False, cmap=cmap)
ax.grid(False)
```

```python
dataroi = data.copy()
dataroi = data.loc[(data['ga.roi'] > -50000)]
dataroi.loc[:,'invbin'] = "none"
rows = (dataroi['initialInvestment'] < 400000)
dataroi.loc[rows, 'invbin'] = "a i<400k"
rows = ((dataroi['initialInvestment'] > 400000) & (dataroi['initialInvestment'] < 800000))
dataroi.loc[rows, 'invbin'] = "b 400k<i<800k"
rows = ((dataroi['initialInvestment'] > 800000) & (dataroi['initialInvestment'] < 1200000))
dataroi.loc[rows, 'invbin'] = "c 800k<i<1.2m"
rows = (dataroi['initialInvestment'] > 1200000)
dataroi.loc[rows, 'invbin'] = "d i>1.2m"
dataroi
```

```python

grid = sns.FacetGrid(dataroi, col="ga.roiyear", row="invbin", size=4)
grid.map(plt.scatter,  "powerProduced", "ga.roi")
```
